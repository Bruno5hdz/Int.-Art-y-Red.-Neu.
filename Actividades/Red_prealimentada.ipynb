```python
import numpy as np
from tensorflow.keras.datasets import mnist
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
(train_data, train_labels), (test_data, test_labels) = mnist.load_data()
```

```python
train_data = train_data.reshape((train_data.shape[0], -1)).astype('float32') / 255.0
test_data = test_data.reshape((test_data.shape[0], -1)).astype('float32') / 255.0
```

```python
model = MLPClassifier(
    hidden_layer_sizes=(128, 64),
    activation='relu',
    solver='adam',
    learning_rate_init=0.001,
    max_iter=20,
    random_state=42
)
```

```python
model.fit(train_data, train_labels)
```

```python
test_predictions = model.predict(test_data)
accuracy = accuracy_score(test_labels, test_predictions)
print(f'Exactitud del modelo en el conjunto de prueba: {accuracy:.4f}')
```

```python
indices = np.random.choice(len(test_data), 5, replace=False)

for i, idx in enumerate(indices):
    img = test_data[idx].reshape(28, 28)
    pred = model.predict([test_data[idx]])[0]
    actual = test_labels[idx]

    plt.imshow(img, cmap='gray')
    plt.title(f'Predicción: {pred} | Valor real: {actual}')
    plt.axis('off')
    plt.show()
```
